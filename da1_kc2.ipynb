# Import neccessary python packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
```

## Read in Kansa City Housing Prices .csv dataset
- dataset found at Kaggle : https://www.kaggle.com/datasets/astronautelvis/kc-house-data

```python
df = pd.read_csv('/Users/johnchristophersloan/Data Analysis Knowledge Check/assets/kc_final.csv')
df.head(10)
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(x='sqft_living',
                y='price',
                data=df,
                hue='zipcode',
                s=120)

plt.title('Housing Prices with relation to Square Footage and Zipcode')

plt.show()
```
